```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):# x is title of the movie
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id# based on id it seacht and get request from there
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))# get rquest from url
    data_json = response.json()# get response in form of json
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```
